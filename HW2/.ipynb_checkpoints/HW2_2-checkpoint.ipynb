```python
import pandas as pd
result = pd.read_table("../HW2/data_puntuation.CSV",encoding="utf-8")
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


```

```python
from datetime import datetime

date2=[]
for i in range(0, len(result.Date)):
    date2.append(datetime.strptime(str(result.Date[i]) , '%d/%m/%Y'))
    
    
result['date2']=date2  

result = result.sort(['date2'], ascending=[1])
```

## Comparing Different Methods

### Sum of methods

Firstly we want to compare how important are the different topics, to do so we sum the ranking obtained by each topic in all of the documents.

```python

fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(111)
## the data
N = 4
COUNT_total =[result.Eco_Count.sum(),result.Legal_Count.sum(),result.Military_Count.sum(),result.Religion_Count.sum()]
TF_IDF_total =[result.Eco_TD.sum(),result.Legal_TD.sum(),result.Military_TD.sum(),result.Religion_TD.sum()]
## necessary variables
ind = np.arange(N)                # the x locations for the groups
width = 0.35                      # the width of the bars
## the bars
rects1 = ax.bar(ind, COUNT_total, width,color='#253494' )

rects2 = ax.bar(ind+width, TF_IDF_total, width,color='#7fcdbb')
# axes and labels
ax.set_xlim(-width,len(ind)+width)
ax.set_ylabel('Scores')
ax.set_title('Scores by method and Dictionary',fontsize=18)
xTickMarks = ['Economy','Legal','Military','Religion']
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=-45, fontsize=10)

## add a legend
ax.legend( (rects1[0], rects2[0]), ('Count', 'TF_IDF') )
plt.show()
```

We can observe in this graph how through both the count method or the TF-IDF method the topic which is more seen in the documents are those related with the Economy. 

### Evolution of the themes through the years


In the following graph we can see the importance of a topic depending on the year and method. We can see that more or less the evolution is the same using either of the two methods. Althought TD-IDF scores higher the less dominant categories (Legal, Military, Religion).

```python
year=[]
for i in range(0, len(result.date2)):
    year.append(result.date2[i].year)
    
    
result['year']=year  

summCount_Year=result.groupby('year')['Eco_Count','Legal_Count','Military_Count','Religion_Count'].sum()
summTDIDF_Year=result.groupby('year')['Eco_TD','Legal_TD','Military_TD','Religion_TD'].sum()
```

```python
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(111)
N = 11

ind = np.arange(N)                # the x locations for the groups
width = 0.2                     # the width of the bars
## the bars
rects1 = ax.bar(ind, summCount_Year.Eco_Count, width,color='#3288bd' )
rects2 = ax.bar(ind+width, summCount_Year.Legal_Count, width,color='#80cdc1' )
rects3 = ax.bar(ind+(2*width), summCount_Year.Military_Count, width,color='#f46d43' )
rects4 = ax.bar(ind+(3*width), summCount_Year.Religion_Count, width,color='#ca0020' )
ax.set_xlim(-width,len(ind)+width)
xTickMarks = summCount_Year.index
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=-45, fontsize=10)
ax.set_title('Scores by Year and Dictionary with the *Count* method',fontsize=18)

ax.set_ylabel('Scores',fontsize=12)
ax.legend( (rects1[0], rects2[0], rects3[0], rects4[0]), ('Economy', 'Legal','Military','Regilion') )
plt.show()

```

```python
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(111)
N = 11

ind = np.arange(N)                # the x locations for the groups
width = 0.2                     # the width of the bars
## the bars
rects1 = ax.bar(ind, summTDIDF_Year.Eco_TD, width,color='#3288bd' )
rects2 = ax.bar(ind+width, summTDIDF_Year.Legal_TD, width,color='#80cdc1' )
rects3 = ax.bar(ind+(2*width), summTDIDF_Year.Military_TD, width,color='#f46d43' )
rects4 = ax.bar(ind+(3*width), summTDIDF_Year.Religion_TD, width,color='#ca0020' )
ax.set_xlim(-width,len(ind)+width)
xTickMarks = summTDIDF_Year.index
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=-45, fontsize=10)
ax.set_title('Scores by Year and Dictionary with the *TD-IDF* method',fontsize=18)

ax.set_ylabel('Scores',fontsize=12)
ax.legend( (rects1[0], rects2[0], rects3[0], rects4[0]), ('Economy', 'Legal','Military','Regilion') )
plt.show()
```

###Top Documents
Finally we are going to check which topic is most important for each of the documents, and see how documents are distributed throught the topics. That means that for each document we are going to check which category had the higher TD-IDF score. The category that has the highest score will count as 1 document of that dictionary.
As expected the economy topic is the category which the td-idf score is the highest for most of the document. Moreover we can see that military and legal category are very equal in the importance in the documents.

```python
topics=result[['Eco_TD','Legal_TD','Military_TD','Religion_TD']]
topics=topics.reset_index(drop=True)
rowmax = topics.max(axis=1)
maxTopic=np.where(topics.values == rowmax[:,None]) # which is maximum of the 4 categories
maxTopicInd=maxTopic[1]
len(topics.Eco_TD)
rankdocuments=[]
cate=[0,1,2,3]
for cat in cate: # how many documents are more important in each category
    CatDoc=np.where(maxTopicInd==cat)
    CatDoc=map(list, CatDoc)
    rankdocuments.append(len(CatDoc[0]))
```

```python
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot(111)
## the data
N = 4

## necessary variables
ind = np.arange(N)                # the x locations for the groups
width = 0.35                      # the width of the bars
## the bars
rects1 = ax.bar(ind, rankdocuments, width,color='#253494' )

#rects2 = ax.bar(ind+width, TF_IDF_total, width,color='#7fcdbb')
# axes and labels
ax.set_xlim(-width,len(ind)+width)
ax.set_ylabel('Number of documents')
ax.set_title('Number of documents in Each category',fontsize=18)
xTickMarks = ['Economy','Legal','Military','Religion']
ax.set_xticks(ind+0.2)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=-45, fontsize=10)

## add a legend
plt.show()
```
